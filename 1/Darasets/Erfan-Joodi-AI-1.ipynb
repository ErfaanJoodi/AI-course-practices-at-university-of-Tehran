# Linear regression

for this part, we're gonna work on diamonds dataset

first of all let's see what's in the dataset

```python
diamonds= pd.read_csv('diamonds.csv')
diamonds.head()
```

so let's make a copy of dataset and work on it to keep the original dataset without any changes

```python
df= diamonds.copy()
```

in this part we're gonna see the features and type of them

```python
df.info()
```

let's see how many null value any feature has

```python
df.isna().sum()
```

now let's see some statistics information about dataset(don't forget that this part is for numerical features only)

```python
df.describe()
```

let's see the number of diamonds that we have in dataset

```python
len(df)
```

as you can see, the number of data and max of "Unnamed: 0" column is same, so that column actually was index, so we can replace it with index

```python
df.index= df['Unnamed: 0'].values.tolist()
df.drop(columns= ['Unnamed: 0'], inplace= True)
df.head()
```

we want to see the correlation between features, but as you can see, there is feature that has "object" type, so we should do label encoding to give those features numbers

```python
label_encoder= LabelEncoder()
for column in df.columns:
    if df[column].dtype == 'object':
        df[column]= label_encoder.fit_transform(df[column])
df.head()
```

now we can see the correlation between features

```python
df.corr()
```

let's make a heatmap of correlation between features in dataset

```python
plt.style.use('fivethirtyeight')
fig, ax= plt.subplots(figsize= (8, 4), dpi= 848)
heatmap= sns.heatmap(df.corr(), ax= ax, vmin= -1, vmax= +1, annot= True, annot_kws={"fontsize": 8}, fmt= '.1f', linecolor= 'white', linewidths= 2, cmap= 'RdBu')
cbar = heatmap.collections[0].colorbar
cbar.ax.tick_params(labelsize=8)
ax.set_yticklabels(df.columns, fontsize= 8)
ax.set_xticklabels(df.columns, fontsize= 8, rotation= 0)
ax.set_title('"Heatmap of correlation between features in diamonds dataset"', fontsize= 10);
```

now by considering the heatmap result, let's make the linear regression model with "price" column as target and "carat" column

```python
X= df[['carat']]
y= df['price']
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.15, random_state= 48)
model= LinearRegression()
model.fit(X_train, y_train)
y_pred= model.predict(X_test)
mse= mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```

```python
plt.style.use('fivethirtyeight')
fig, ax= plt.subplots(figsize= (10, 4), dpi= 848)
ax.scatter(X_train, y_train, color='blue', label='Training data')
ax.scatter(X_test, y_test, color='green', label='Testing data')
ax.plot(X_test, y_pred, color='red', linewidth=2, label='Regression line')
ax.set_title('"Linear regression model"', fontsize= 12)
ax.set_xlabel('Carat', fontsize= 10)
ax.set_ylabel('Price', fontsize= 10)
ax.legend(fontsize= 8)
ax.set_xticks([0, 1, 2, 3, 4, 5])
ax.set_xticklabels([0, 1, 2, 3, 4, 5], fontsize= 8)
ax.set_yticks([0, 5000, 10000, 15000, 20000])
ax.set_yticklabels([0, 5000, 10000, 15000, 20000], fontsize= 8);
```
